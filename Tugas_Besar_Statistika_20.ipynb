* **Korelasi dan Scatter Plot**

> Data yang digunakan pada scatter plot ialah Total_Trans_Amt, Total_Trans_Ct, dan Marital_Status. **Total_Trans_Amt merupakan variabel Jumlah Transaksi Total, Total_Trans_Ct merupakan variabel Total Jumlah Transaksi dan Marital_Status merupakan variabel Status Pernikahan (Menikah, Lajang, Bercerai, Tidak Diketahui)**

> Data yang digunakan pada plot koordinat cartesius ialah Total_Trans_Amt, Total_Amt_Chng_Q4_Q1, dan Gender. **Total_Trans_Amt merupakan variabel Jumlah Transaksi Total, Total_Amt_Chng_Q4_Q1 merupakan variabel perubahan Jumlah Transaksi (Q4 sampai Q1).**

```
# Menampilkan Scatter plot antara variabel Total_Trans_Amt, Total_Trans_Ct, dan Marital_Status
ggplot(bank, aes(x = Total_Trans_Amt, y = Total_Trans_Ct, color = Marital_Status)) +
  geom_point(alpha = .3) +
  scale_x_log10() +
  scale_y_log10() +
  facet_wrap(~ Marital_Status, nrow = 2)
```

Dari scatter plot di atas dapat disimpulkan bahwa total jumlah yang melakukan transaksi terpadat berada pada jumlah total traksaksi 30-100 dan yang lebih dominan dalam melakukan transaksi yang berstatus Married (sudah menikah).

```
# Menentukan nilai korelasi pearson antara variabel Total_Trans_Amt dan Total_Trans_Ct pada data BankChurnes
cor(bank$Total_Trans_Amt, bank$Total_Trans_Ct, method = "pearson")
```

Jika kita hitung korelasi dari Total_Trans_Amt (Jumlah Total Transaksi) dan Total_Trans_Ct (Total Jumlah Transaksi) menghasilkan korelasi yang bernilai 0.807192034651435 yang tampak mempunyai korelasi positif kuat.

Untuk mengetahui ada tidaknya korelasi antara Total_Trans_Ct dan Total_Trans_Amt, kita dapat menggunakan correlation test.

```
# Korelasi test
cor.test(bank$Total_Trans_Amt, bank$Total_Trans_Ct, method = "pearson")
```

t = 137.6, df = 10125 (df adalah singkatan dari degree of freedom atau derajat kebebasan), p-value < 2.2e-16 (0.00000000000000022), hipotesis alternatif : korelasi sebenarnya tidak sama dengan 0 di Interval kepercayaan 95 persen: 0.8002971 0.8138735

Berdasarkan hasil diatas diperoleh *p-value* < 5%, atau dapat diperoleh bahwa tingkat kesalahan H1 lebih kecil dibandingkan tingkat kesalahan Ho.

Kesimpulan: Tolak Ho/Terima H1, sehingga terdapat korelasi antara variabel *Total_Trans_Amt* dan *Total_Trans_Ct*.

```
# Membuat plot 3 variabel dalam koordinat Cartesius
ggplot(bank, aes(x = Total_Trans_Amt, y = Total_Amt_Chng_Q4_Q1, color = Gender)) +
  geom_point(alpha = .5)
```

Dari scatter plot di atas dapat disimpulkan bahwa perubahan jumlah yang melakukan transaksi terpadat berada pada perubahan hitungan traksaksi 0-1 dan yang lebih dominan dalam melakukan perubahan transaksi adalah Female.

```
cor(bank$Total_Trans_Amt, bank$Total_Amt_Chng_Q4_Q1, method = "pearson")
```

Jika kita hitung korelasi dari Total_Amt_Chng_Q4_Q1 (Perubahan Jumlah Transaksi) dan Total_Trans_Amt (Jumlah Total Transaksi) menghasilkan korelasi yang bernilai 0.0396775923718641 yang dimana tidak memiliki suatu korelasi.

Untuk mengetahui nilai korelasi antara Total_Amt_Chng_Q4_Q1 dan Total_Trans_Amt, kita dapat menggunakan correlation test.

```
cor.test(bank$Total_Trans_Amt, bank$Total_Amt_Chng_Q4_Q1, method = "pearson")
```

t = 3.9956, df = 10125 (df adalah singkatan dari *degree of freedom* atau derajat kebebasan), p-value = 6.498e-05, hipotesis alternatif : korelasi sebenarnya tidak sama dengan 0 di Interval kepercayaan 95 persen: 0.02021644 0.05910869

Berdasarkan hasil diatas diperoleh *p-value* < 5%, atau dapat diperoleh bahwa tingkat kesalahan H1 lebih kecil dibandingkan tingkat kesalahan Ho.

Kesimpulan: Tolak Ho/Terima H1, sehingga terdapat korelasi antara variabel *Total_Trans_Amt* dan *Total_Amt_Chng_Q4_Q1*.
